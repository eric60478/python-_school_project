```python
def get_txt_fitz(PlaceTxt,PdfLink,TxtName,filename='total'):
    '''PlaceTxt選擇放置的資料夾的位置,PdfLink選擇檔案,TxtName創造放置資料夾的名字,filename放置檔案的pdf名稱'''
    os.chdir(PlaceTxt)
    if  not os.path.isdir(PlaceTxt+'\\'+TxtName):
        os.mkdir(PlaceTxt+'\\'+TxtName)
    file = fitz.open(PdfLink)
    txt=open(PlaceTxt+'\\'+TxtName+'\\'+filename+'.txt','w',encoding='utf-8')
    for page in file.pages():
        text=page.getText()
        txt=open(PlaceTxt+'\\'+TxtName+'\\'+filename+'.txt','a',encoding='utf-8')
        print(text)
        print("************************")
        txt.write(text)
        txt.close()
```

```python
import os
import fitz
import time
def transfer_allpdf_getdetail():
    wherepdf=input("wherepdf")
    PlaceTxt=input("檔案資料夾要放哪")
    filename=input("此資料夾要叫的名字")
    files= os.listdir(wherepdf)
    totalindexs=''
    for name in files:
        PdfLink=wherepdf+"\\"+name
        name=name.strip("pdf")
        start=time.time()
        get_txt_fitz(PlaceTxt,PdfLink,filename,name)
        end=time.time()
        processtime=end-start
        pdflen=open(PlaceTxt+'\\'+filename+'\\'+name+'.txt','r',encoding='utf-8')
        pdflens=len(pdflen.read())
        pdflen.close()
        totalindex=name+"Time:  "+str(processtime)+"Len:  "+str(pdflens)+'\n'
        totalindexs=totalindexs+totalindex
    file_of_index=open(PlaceTxt+'\\'+'PDF_process_detail.txt','w',encoding='utf-8')
    file_of_index.writelines(totalindexs)
transfer_allpdf_getdetail()
```
