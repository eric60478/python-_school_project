<font color=black size=10 face=雅黑>**取出json(主要要看格式有些不符)或html的資料分析**</font>

```python
from bs4 import BeautifulSoup
import json
title_tag=[]
body=[]
jsonData=[]
try :
    soup = BeautifulSoup(data, 'html.parser')     
    title_tag = soup.html.find_all("title")
    body = soup.html.find_all("body")
except:
    jsonData=json.loads(data)
```

<font color=black size=10 face=雅黑>**進行標題和內容的切分可以相互對照**</font>

```python
title_tag_cut=[]
body_cut=[]
jsonData_cut=[]
temp=""
temp2=""
if title_tag:
    for i in title_tag:
        for char in str(i):
            if '\u4e00' <= char <= '\u9fa5':
                temp=temp+char                
        title_tag_cut.append(temp)
        temp=""
if body:
    for i in body:
        for char in str(i):
            if '\u4e00' <= char <= '\u9fa5':                
                temp=temp+char                
        body_cut.append(temp)
        temp=""
if  jsonData:
    jsonData_key=list( jsonData[0].keys())
    for i in  jsonData:
        for j in  jsonData_key:
            for char in str(i[str(j)]):
                if '\u4e00' <= char <= '\u9fa5':                
                    temp2=temp2+char  
        temp=temp+temp2
        temp2=""
        jsonData_cut.append(temp)
        temp=""
```

<font color=black size=10 face=雅黑>**html標題切分title**</font>

```python
for string in title_tag_cut:
    for word in jieba.cut(str(string)):
        print(word,end='  ')
    print("***")
```

<font color=black size=10 face=雅黑>**html標題切分內容**</font>

```python
for string in body_cut:
    for word in jieba.cut(str(string)):
        print(word,end='  ')
    print("***")
```

<font color=black size=10 face=雅黑>**jason切分依每組資訊**</font>

```python
for string in jsonData_cut:
    for word in jieba.cut(str(string)):
        print(word,end='  ')
    print("***")
```
